```python
class Animal:    
    
    def __init__(self, name, color):
        
        self.name = name
        self.color = color
        self.movement = 'walk'
    # non inserito metodo __str__ inserito nelle diverse classi "child"
    def greet(self):
        
        your_name = input('What\'s your name? ')
        return f'Hello {your_name}, my name is {self.name} and i\'m an amazing {self.color} {self.species}. Nice to meet you!'
    
    def move(self):
        
        return f'I am {self.movement}ing'

class Sheep(Animal):
    
    def __init__(self, name, color):
        
        super().__init__(name, color)
        self.species = 'Sheep'
        self.number_of_legs = 4
        
    def __str__(self):
        
        return f'''
       Name: {self.name}
    Species: {self.species}
      Color: {self.color}
       Legs: {self.number_of_legs}'''
    
    def speak(self):
        
        return self.name + ' says beeeee'
    
class Wolf(Animal):
    
    def __init__(self, name, color):
        
        super().__init__(name, color)
        self.species = 'Wolf'
        self.number_of_legs = 4
        
    def __str__(self):
        
        return f'''
       Name: {self.name}
    Species: {self.species}
      Color: {self.color}
       Legs: {self.number_of_legs}'''
    
    def speak(self):
        
        return self.name + ' says uuuuuuu'
    # override della greet function che ritorna la greet function della cls Animal tramite super() + greet dell'istanza
    def greet(self):
        
        wolf_line = " Don't worry i am not a werewolf!"
        return super().greet() + wolf_line
        
class Snake(Animal):    
    
    def __init__(self, name, color):
        
        super().__init__(name, color)
        self.species = 'Snake'
        self.number_of_legs = 0
        #override dell'attributo self.movement che viene richiamato dalla funzione del suo "parent" ma con valore del "child"
        self.movement = 'crawl'

    def __str__(self):
        
        return f'''
       Name: {self.name}
    Species: {self.species}
      Color: {self.color}
       Legs: {self.number_of_legs}'''
    
    def speak(self):
        
        return self.name + ' says ssssssss'

class Parrot(Animal):    
    #added new init parameter only for class Parrot
    def __init__(self, name, color, wings_color):
        
        super().__init__(name, color)
        self.species = 'Parrot'
        self.number_of_legs = 2
        self.wings_color = wings_color
        self.movement = 'fly'

    def __str__(self):
        # return super().__str__() + parametro aggiunto di parrot
        return f'''
       Name: {self.name}
    Species: {self.species}
      Color: {self.color}
Wings Color: {self.wings_color}
       Legs: {self.number_of_legs}'''        
        
    def speak(self):
        
        return self.name + ' says ciaooooo'
    
bella = Sheep('Bella','White')
alberto = Wolf('Alberto','Grey')
aoda = Snake('Aoda','Blue')
cocorito = Parrot('Cocorito','Red and White', 'Yellow')

my_animals = [bella,alberto,aoda,cocorito]
```

```python
# grazie al metodo __str__ quando vado a printare l'istanza ottengo qualcosa
for i in my_animals:
    print(i)
```

```python
# 
def animal_speak(animal):
    print(animal.speak())
```

```python
print('________________________________')
for i in my_animals:
    print(i,'\n')
    animal_speak(i)
    print('\n',i.move(),'\n________________________________')
```
