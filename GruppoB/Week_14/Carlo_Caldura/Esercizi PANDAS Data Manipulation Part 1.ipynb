#### Exercise 12
Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
import numpy as np
```

```python
np.random.seed(123)
arr1 = np.round(np.random.normal(size=10), 2)
arr2 = np.round(np.random.normal(size=10), 2)
arr3 = np.round(np.random.normal(size=10), 2)
arr4 = np.round(np.random.normal(size=10), 2)

arr1[np.random.choice(10, 2)] = np.nan
arr2[np.random.choice(10, 5)] = np.nan
arr3[np.random.choice(10, 1)] = np.nan
arr4[np.random.choice(10, 2)] = np.nan

df = pd.DataFrame({'A':arr1, 'B':arr2, 'C':arr3, 'D':arr4})
df.style.highlight_null(null_color='yellow')
```

#### Exercise 13
Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframe’s specific columns with different colors.

```python
np.random.seed(123)

arr1 = np.random.randint(0,10,10)
arr2 = np.random.randint(0,10,10)
arr3 = np.random.randint(0,10,10)
arr4 = np.random.randint(0,10,10)

df = pd.DataFrame({'A':arr1, 'B':arr2, 'C':arr3, 'D':arr4})

def highlight(value):
    
    df = value.copy()
    df.loc[:, ['A', 'D']] = 'background-color: red'
    df.loc[:, ['B', 'C']] = 'background-color: green'
    return df

df.style.apply(highlight, axis=None)
```
